```python
import requests
from bs4 import BeautifulSoup
import pandas as pd



c=[]

page = requests.get('https://www.worldometers.info/coronavirus/')
soup = BeautifulSoup(page.content, 'html.parser')

table = soup.find('table',id = 'main_table_countries_today')
tbodys = table.find_all('tbody')

tbodys = tbodys[0]
#print(tbodys)

for row in tbodys.find_all('tr', {'class': ''}):
    for cell in row.find_all('td'):
        c.append((cell.text))

        
        
#print(c)
d =[]
country=[]
totalcase=[]
newcases=[]
totaldeaths=[]
newdeaths=[]
totalrecover=[]
activecases=[]
seriouscritical=[]
totaltest=[]
continental=[]
i=0
j=0

for i in range (i, len(c),13):
    d.append(c[i:i+13])

#print(d)

for j in range (j,len(d),1):
    country.append(d[j][0])
    totalcase.append(d[j][1])    
    newcases.append(d[j][2])
    totaldeaths.append(d[j][3])
    newdeaths.append(d[j][4])
    totalrecover.append(d[j][5])
    activecases.append(d[j][6])
    seriouscritical.append(d[j][7])
    totaltest.append(d[j][10])
    continental.append(d[j][12])
# print(country)
  
# print(totalcase)


       
df = pd.DataFrame({
    'Country': country,
    'Total_Case': totalcase,
    'New_Cases' : newcases,
    'Total_Deaths': totaldeaths,
    'New_Deaths' : newdeaths,
    'Total_Recover' : totalrecover,
    'Active_Cases' : activecases,
    'Serious_Critical' : seriouscritical,
    'Total_Test' : totaltest,
    'Contitental' : continental
    
    
})
print(df)


#df.to_csv('corona_info.csv',index=False)
```
